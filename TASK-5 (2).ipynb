```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To make plots look better
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load the Titanic train.csv file
df = pd.read_csv('train.csv')

# Show first 5 rows
df.head()
```

```python
# General information about the dataset
df.info()

# Summary statistics
df.describe()

# Check missing values
df.isnull().sum()

```

```python
# Value counts for Survived column
print(df['Survived'].value_counts())

# Value counts for Sex
print(df['Sex'].value_counts())

# Value counts for Passenger Class (Pclass)
print(df['Pclass'].value_counts())
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot of Age vs Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survived')
plt.show()
```

```python
# Scatterplot of Age and Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age (colored by Survival)')
plt.show()
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
# Sex vs Survival
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survival')
plt.show()
```

```python
# Passenger Class vs Survival
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Passenger Class vs Survival')
plt.show()
```

```python
- Most passengers are young (20-40 years).
- Very few elderly passengers.
- Female passengers had a much higher survival rate than male passengers.
- First-class passengers (Pclass = 1) had higher survival rates compared to second and third class.
- Passengers who paid higher fares were more likely to survive.
- Younger passengers (especially children) had a higher survival rate.
- The dataset has missing values in 'Age' and 'Cabin' columns.
- Feature engineering may be necessary for better predictive modeling.
```
